# Dataloader class

```python
def get_transform(train):
    transform_list = []
    transform_list.append(transforms.Resize(256))
    transform_list.append(transforms.CenterCrop(224))
    if train:
        transform_list.append(transforms.RandomHorizontalFlip())
    transform_list.append(transforms.ToTensor())
    transform_list.append(transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]))

    return transforms.Compose(transform_list)
```

```python
class CustomFruitDataset(Dataset):
    def __init__(self, csv_path, transforms=None):
        self.csv = pd.read_csv(csv_path)
        self.transforms = transforms

    def __len__(self):
        return len(self.csv)

    def __getitem__(self, idx):
        sample_row = self.csv.iloc[idx]
        label = sample_row["category"]
        
        image = Image.open(os.path.join(base_path, sample_row["image:FILE"])).convert('RGB')

        if self.transforms:
            image = self.transforms(image)

        return image, label
```

```python
def denormalize(image_tensor):
    mean = torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1)
    std = torch.tensor([0.229, 0.224, 0.225]).view(3, 1, 1)
    image_tensor = image_tensor * std + mean
    return image_tensor.clamp(0, 1)
```

```python
custom_ds = CustomFruitDataset(
    csv_path=os.path.join(base_path, "train.csv"),
    transforms=get_transform(train=True)
)

image, label = custom_ds[0]
image_denormalize = denormalize(image)
image_denormalize = image_denormalize.numpy().transpose((1, 2, 0))


fig, axes = plt.subplots(1, 2, figsize=(15, 8))

axes[0].imshow(image.permute(1, 2, 0))
axes[0].set_title("Before denormalization")

axes[1].imshow(image_denormalize)
axes[1].set_title("After denormalization")

fig.suptitle(f"Fruit name: {class_name[label]}")
plt.show()
```

# Model and dataloader

```python
def get_model(num_classes):
    model = torchvision.models.resnet50(weights="DEFAULT")
    try:
        model.fc = nn.Linear(model.fc.in_features, num_classes)
    except AttributeError:
        model.classifier[-1] = nn.Linear(model.classifier[-1].in_features, num_classes)
    return model
```

```python
model = get_model(len(class_name)).to(device)
summary(model, (3, 256, 256))
```

```python
dataset_train = CustomFruitDataset(
    csv_path=os.path.join(base_path, "train.csv"),
    transforms=get_transform(train=True)
)
dataset_valid = CustomFruitDataset(
    csv_path=os.path.join(base_path, "val.csv"),
    transforms=get_transform(train=False)
)
```

```python
train_data_loader = torch.utils.data.DataLoader(
    dataset_train,
    batch_size=64,
    shuffle=True,
)

val_data_loader = torch.utils.data.DataLoader(
    dataset_valid,
    batch_size=64,
    shuffle=False,
)
```

# Training

```python
model = get_model(len(class_name)).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
num_epochs = 20
best_accuracy = 0.0
accuracy = Accuracy(task='multiclass', num_classes=len(class_name)).to(device)
```

```python
for epoch in range(num_epochs):
    model.train()
    train_loss = 0.0
    train_loop = tqdm(train_data_loader, desc=f'Epoch {epoch}')
    for inputs, labels in train_loop:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        train_loss += loss.item()
        train_loop.set_postfix(loss=loss.item())

    train_loss /= len(train_data_loader)

    model.eval()
    val_loss = 0.0
    val_accuracy = 0.0
    val_loop = tqdm(val_data_loader, desc=f"Validation {epoch}")
    with torch.no_grad():
        for inputs, labels in val_loop:
            inputs, labels = inputs.to(device), labels.to(device)
            
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_loss += loss.item()
            val_accuracy += accuracy(outputs, labels).item()

    val_loss /= len(val_data_loader)
    val_accuracy /= len(val_data_loader)

    print(f'Epoch {epoch}: Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | Val Accuracy: {val_accuracy:.4f}')

    if val_accuracy > best_accuracy:
        best_accuracy = val_accuracy
        torch.save(model.state_dict(), 'best_model.pth')
        print(f"ModÃ¨le sauvegardÃ© avec une prÃ©cision de validation de {best_accuracy:.4f}")
```
